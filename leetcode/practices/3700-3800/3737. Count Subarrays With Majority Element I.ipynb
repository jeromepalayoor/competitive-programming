```python
class Solution(object):
    def countMajoritySubarrays(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        n = len(nums)
        pref = [0] * (n+1)
        for i in range(n):
            pref[i+1] = pref[i] + (1 if nums[i] == target else 0)
        
        count = 0
        for i in range(n):
            for j in range(i, n):
                if (pref[j+1] - pref[i]) * 2 > j - i + 1:
                    count += 1

        return count
        

s = Solution()
print(s.countMajoritySubarrays([1, 2, 2, 3], 2))
print(s.countMajoritySubarrays([1, 1, 1, 1], 1))
print(s.countMajoritySubarrays([1, 2, 3], 4))
```
