```python
class Solution(object):
    def maximumLength(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        count = {}
        for n in nums:
            if n in count:
                count[n] += 1
            else:
                count[n] = 1

        best = 1
        if 1 in count:
            count_1 = count[1]
            if count_1 % 2 == 0:
                count_1 -= 1
            best = max(count_1, best)

        for n in count.keys():
            if n == 1 or count[n] == 1:
                continue
            if n**2 not in count:
                continue
            curr = 2
            while n ** 2 in count:
                n = n**2
                if count[n] < 2:
                    curr += 1
                    break
                if n**2 in count:
                    curr += 2
                else:
                    curr += 1
                    break
            best = max(curr, best)

        return best
```

```python
s = Solution()
print(s.maximumLength([5, 4, 1, 2, 2]))
print(s.maximumLength([1, 2, 3, 4]))
```
