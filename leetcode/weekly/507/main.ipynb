```python
class Solution(object):
    def maxDistance(self, moves):
        """
        :type moves: str
        :rtype: int
        """
        u, d, l, r, x = 0, 0, 0, 0, 0
        for m in moves:
            if m == 'U': u += 1
            if m == 'D': d += 1
            if m == 'L': l += 1
            if m == 'R': r += 1
            if m == '_': x += 1
        return abs(u-d) + abs(l-r) + x

s = Solution()
print(s.maxDistance('L_D_'))
print(s.maxDistance('U_R_'))
```

```python
class Solution(object):
    def countValidSubarrays(self, nums, x):
        """
        :type nums: List[int]
        :type x: int
        :rtype: int
        """
        total = 0
        for i in range(len(nums)):
            s = 0
            for j in range(i, len(nums)):
                s += nums[j]
                a = str(s)
                if a[0] == a[-1] and a[0] == str(x):
                    total += 1
        return total
        

s = Solution()
print(s.countValidSubarrays([1,100,1], 1))
print(s.countValidSubarrays([1], 2))
```
