```python
class Solution:
    def createGrid(self, m: int, n: int) -> list[str]:
        if m == 1: return ['.'*n]
        if n == 1: return ['.' for _ in range(m)]
        grid = ['.'*n]
        for _ in range(m-1):
            grid.append('#'*(n-1)+'.')
        return grid

s = Solution()
print(s.createGrid(2, 3))
```

```python
class Solution:
    def minLights(self, lights: list[int]) -> int:
        p = {}
        n = len(lights)
        for i in range(n):
            p[i] = -1
        for i, v in enumerate(lights):
            if v != 0:
                p[max(0, i-v)] = min(n, i+v)
                if i-v+1 < n:
                    p[max(0, i-v+1)] = min(n, i+v)
                if i-v+2 < n:
                    p[max(0, i-v+2)] = min(n, i+v)
        m = 0
        i = 0
        safe = -1
        while i < n:
            if i <= safe:
                if p[i] != -1:
                    safe = max(p[i], safe)
                i += 1
                continue
            if p[i] == -1:
                m += 1
                i += 3
                continue
            safe = p[i]
            i += 1
        print(p)
        return m
    
s = Solution()
print(s.minLights([0, 0, 0, 0]))
print()
print(s.minLights([0, 0, 0, 2, 0]))
print()
print(s.minLights([2, 1, 0]))
print()
print(s.minLights([2, 0, 0]))
print()
print(s.minLights([3,1,0,4,5]))
print()
print(s.minLights([3,0,0,5,0,0,1,0,0,0]))
```

```python
class Solution(object):
    def finishTime(self, n, edges, baseTime):
        """
        :type n: int
        :type edges: List[List[int]]
        :type baseTime: List[int]
        :rtype: int
        """
        nodes = {}

        for i in range(n):
            nodes[i] = {
                'base': baseTime[i],
                'childs': [],
                'parent': None,
                'leaf': True
            }


        for i in range(n-1):
            u, v = edges[i]
            nodes[u]['childs'].append(v)
            nodes[v]['parent'] = u
            nodes[u]['leaf'] = False

        def calc(node):
            if nodes[node]['leaf']:
                return nodes[node]['base']
            earliest, latest = 1e10, -1
            for c in nodes[node]['childs']:
                v = calc(c)
                earliest = min(earliest, v)
                latest = max(latest, v)
            return latest + (latest - earliest) + nodes[node]['base']

        return calc(0)
    

s = Solution()
print(s.finishTime(3, [[0,1],[1,2]], [9,5,3]))
print()
print(s.finishTime(3, [[0,1],[0,2]], [4,7,6]))
print()
print(s.finishTime(4, [[0,1],[0,2],[2,3]], [5,8,2,1]))
print()
        
```
